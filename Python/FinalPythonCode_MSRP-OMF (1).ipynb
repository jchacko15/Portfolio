## Target Variable: MSRP

# Data Cleaning and Preprocessing

```python
car.info()
#there are a total of 11914 rows but: 
#'Engine Fuel Type' shows 11911 
#'Engine HP' shows 11845
#'Engine Cylinders' shows 11884
#'Number of Doors' shows 11908
#'Market Category' shows 8172
```

## Dropping Useless Columns

```python
cols = [ 'Model', 'Market Category']
car = car.drop(cols, axis=1)
```

```python
car.info()
```

## Filling Missing Values

```python
print('Missing Engine Fuel Type count:',car['Engine Fuel Type'].isna().sum())
print('Missing Engine HP count:',car['Engine HP'].isna().sum())
print('Missing Engine Cylinders count:',car['Engine Cylinders'].isna().sum())
print('Missing Number of Doors count:',car['Number of Doors'].isna().sum())
```

```python
#Enginge fuel type
car.fillna(value={'Engine Fuel Type':car['Engine Fuel Type'].mode()[0]},
               inplace=True)
```

```python
#Engine HP
car.fillna(value={'Engine HP':car['Engine HP'].mean()},inplace=True)
```

```python
#engine cylinders
car.fillna(value={'Engine Cylinders':car['Engine Cylinders'].mean()},
               inplace=True)
```

```python
#number of doors
car.fillna(value={'Number of Doors':car['Number of Doors'].mode()[0]},
               inplace=True)
```

```python
#check
car.isna().sum()>0
```

## Encoding Categorical Variables

```python
car.info()
```

```python
car_dummies = pd.get_dummies(car)
#dummies: Make, Engine Fuel Type, Transmission Type, Driven_Wheels, Vehicle Size, Vehicle Style 
```

```python
car_dummies.info()
```

```python
car_dummies.drop_duplicates(inplace=True)
```

# Descriptive Analysis

```python
cols = ['MSRP']
car_dummies = car_dummies[cols + [c for c in car_dummies.columns if c not in cols]]
```

```python
car_dummies.describe()
```

# Main Analysis

## Problem Type: Regression

### features and target sets

```python
X = car_dummies.iloc[:,1:] 
Y = car_dummies['MSRP']
```

```python
#check
print(X.info())
print(Y.head(1))
```

### train test split: 80/20

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                                 test_size=.2,
                                                 random_state=0)
```

```python
print(car_dummies.shape)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# Knn Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
#define the model
knn_reg3 = KNeighborsRegressor(n_neighbors=3)
knn_reg4 = KNeighborsRegressor(n_neighbors=4)
knn_reg5 = KNeighborsRegressor(n_neighbors=5)
knn_reg35 = KNeighborsRegressor(n_neighbors=35)
#fitting/training the model
knn_reg3.fit(X_train,Y_train)
knn_reg4.fit(X_train,Y_train)
knn_reg5.fit(X_train,Y_train)
knn_reg35.fit(X_train,Y_train)
```

```python
# evaluate the performance of the model
# score function
print(knn_reg3.score(X_train,Y_train).round(2))
print(knn_reg4.score(X_train,Y_train).round(2))
print(knn_reg5.score(X_train,Y_train).round(2))
print(knn_reg35.score(X_train,Y_train).round(2))
```

```python
print('knn reg3 R2 on train: {:.2%}'.format(knn_reg3.score(X_train,Y_train)))
print('knn reg3 R2 on test: {:.2%}'.format(knn_reg3.score(X_test,Y_test)),'\n')

print('knn reg4 R2 on train: {:.2%}'.format(knn_reg4.score(X_train,Y_train)))
print('knn reg4 R2 on test: {:.2%}'.format(knn_reg4.score(X_test,Y_test)),'\n')

print('knn reg5 R2 on train: {:.2%}'.format(knn_reg5.score(X_train,Y_train)))
print('knn reg5 R2 on test: {:.2%}'.format(knn_reg5.score(X_test,Y_test)),'\n')

print('knn reg35 R2 on train: {:.2%}'.format(knn_reg35.score(X_train,Y_train)))
print('knn reg35 R2 on test: {:.2%}'.format(knn_reg35.score(X_test,Y_test)),'\n')
```

```python
#looking for highest test score^
#n_neighbors of 3 gives highest test score
```

## compare test set performance: validation curve

```python
# cross validation: 
scores10=cross_val_score(knn_reg3, X_test, Y_test,cv=10)
scores9=cross_val_score(knn_reg3,  X_test, Y_test,cv=9)
scores7=cross_val_score(knn_reg3,  X_test, Y_test,cv=7)
scores6=cross_val_score(knn_reg3,  X_test, Y_test,cv=6)
# test scores in each of the folds
print(scores10)
print(scores9)
print(scores7)
print(scores6)
```

```python
# not better than Knn
```

```python
#test different alphas
# smaller alpha--- more complex model, higher chance of over-fitting
# larger alpha-- less complex (simpler)n model, lower chance of over-fitting
### but, if we keep increasing alpha, model might become too simple,
### and we could have under-fitting
ridge1=Ridge(alpha=.3)
ridge2=Ridge(alpha=.5)
ridge3=Ridge(alpha=.8)
```

```python
# fit the model
ridge1.fit(X_train, Y_train)
ridge2.fit(X_train, Y_train)
ridge3.fit(X_train, Y_train)
```

```python
print('ridge1 R2 on train: {:.2%}'.format(ridge1.score(X_train, Y_train)))
print('ridge1 R2 on test: {:.2%}'.format(ridge1.score(X_test, Y_test)),'\n')

print('ridge2 R2 on train: {:.2%}'.format(ridge2.score(X_train, Y_train)))
print('ridge2 R2 on test: {:.2%}'.format(ridge2.score(X_test, Y_test)),'\n')

print('ridge3 R2 on train: {:.2%}'.format(ridge3.score(X_train, Y_train)))
print('ridge3 R2 on test: {:.2%}'.format(ridge3.score(X_test, Y_test)),'\n')
```

```python
# create features and target sets 
X_Sh_Car, Y_Sh_Car=Car_Sh.iloc[:,1:], Car_Sh['MSRP']
# check my work
display(X_Sh_Car.head(1))
display(Y_Sh_Car.head(1))
```

```python
Xs_train, Xs_test, Ys_train, Ys_test=train_test_split(X_Sh_Car,Y_Sh_Car,
                                                 test_size=.2,
                                                 random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor, export_graphviz
```

## Validation Curve for the Decision Tree

```python
train_scores, test_scores=validation_curve(
    DecisionTreeRegressor(random_state=0),Xs_train,Ys_train,
param_name='max_depth',param_range=[5,6,7,8],cv=5)
```

```python
train_scores.mean(axis=1).round(3)
```

```python
test_scores.mean(axis=1).round(3)
# max_depth 8 is the best value
```

```python
# define and fit the model
dt_reg=DecisionTreeRegressor(random_state=0,max_depth=8)
dt_reg.fit(Xs_train,Ys_train)
```

```python
print('dt_reg on train: {:.2%}'.format(dt_reg.score(Xs_train, Ys_train)))
print('dt_reg on test: {:.2%}'.format(dt_reg.score(Xs_test, Ys_test)))
```
